````python
from langchain.prompts import PromptTemplate
from llms import get_llama

refine_prompt = PromptTemplate.from_template(
    """
    Com base no seguinte resumo de um acórdão do TRE e os trechos do contexto, refine o resumo, incluindo as seguintes informações omitidas no resumo original, mantendo sua estrutura original:

    Resumo original:
    {original}

    Informações omitidas:
    {missing}

    Trechos do contexto:
    {context}

    Resumo refinado:
    """
)
````

````python
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_community.vectorstores import FAISS
from langchain_ollama import OllamaEmbeddings

embeddings_model = OllamaEmbeddings(
    model='nomic-embed-text',
    base_url='http://10.10.0.99:11434',
)
original_document = split_documents(source_file_path)
vectorstore = FAISS.from_documents(documents=original_document, embedding=embeddings_model)
retriever = vectorstore.as_retriever()

llm = get_llama()

stuff_chain = create_stuff_documents_chain(llm, refine_prompt)
retrieval_chain = create_retrieval_chain(retriever, stuff_chain)
````

````python
from langchain_core.messages.ai import AIMessageChunk

def stream_chain(chain, input, print_result=False):
    chunks = []
    for chunk in chain.stream(input):
        if type(chunk) is AIMessageChunk:
            answer = chunk.content
        else:
            answer = chunk.get('answer', '')
        chunks.append(answer)
        if print_result:
            print(answer, end='', flush=True)
    if print_result:
        print()
    result = ''.join(chunks)
    return result
````

````python
chunks = []

for chunk in retrieval_chain.stream({
    'original': processed_file_contents,
    'missing': missing_words,
    'input': 'Comece.',
}):
    answer = chunk.get('answer', '')
    chunks.append(answer)
    print(answer, end='', flush=True)
print()

res = "".join(chunks)
````

````python
prompt = PromptTemplate.from_template("""Encontre sentenças redundantes no texto refinado e liste-as.

Texto refinado:
{refined}

Lista de sentenças redundantes:
""")

redundancy_chain = prompt | llm
lista_de_redundancias = stream_chain(redundancy_chain, {'refined': res}, print_result=True)
````

````python
refine_prompt = PromptTemplate.from_template("""Com base na lista de redundâncias, remova as sentenças redundantes do texto refinado. Responda com o texto refinado sem as sentenças redundantes, e nada mais.

Lista de sentenças redundantes:
{redundancies}

Texto refinado:
{refined}

Texto refinado sem redundâncias:""")
refine_chain = refine_prompt | llm

res = stream_chain(refine_chain, {'redundancies': lista_de_redundancias, 'refined': res}, print_result=True)
````
